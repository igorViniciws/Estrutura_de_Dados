```python
# Carregar o arquivo

import pandas as pd

df = pd.read_excel("dados_ausentes.xlsx")

print(df)
```

```python
# Verificar se existem dados faltando

print(df.isnull())
```

```python
# Contar o número de ausencias por coluna

print(df.isnull().sum())
```

```python
# Excluir a linha dos dados ausentes

df_sem_vazio = df.dropna()
print(df_sem_vazio)
```

```python
# Substituir dados ausentes
# Cidade vazia - desconhecido

df["Cidade"].fillna("Desconhecido", inplace=True)
print(df)
```

```python
# Sustituir dados ausentes
# Substituir "Salário" por 0

df["Salário"].fillna(0, inplace=True)
print(df)
```

```python
# Substituir dados ausentes
# Idade por média

df["Idade"].fillna(df["Idade"].mean(), inplace=True)
print(df)
```

```python
# Corrigir os erros de formatação (minúsculo e maiúsculo)
df["Nome"] = df["Nome"].str.capitalize()
df["Cidade"] = df["Cidade"].str.title()

print(df)
```

```python
# Padronização de nomenclatura
# replace - sustitui o valor
df["Cidade"] = df["Cidade"].str.upper().replace({
                                                    "SÃO PAULO": "SP",
                                                    "RIO DE JANEIRO": "RJ",
                                                    "SALVADOR": "SSA",
                                                    "PORTO ALEGRE": "POA",
                                                })

print(df)
```

```python
# Identificar o tipo de dado em uma coluna
# Tipo String

coluna_string = df.select_dtypes(include=["string", "object"]).columns.to_list()
print(coluna_string)
```

```python
# Trocar o tipo de dado de idade para INT
df["Idade"]= df["Idade"].astype(int)
print(df)
```

```python
# Pega o valor de cima e substituir a ausencia
# df["Idade"] = df["Idade"].ffill()

# Debaixo
# df["Idade"] = df["Idade"].bfill
```

```python
# Criar uma nova coluna: Faixa Etária
df["Faixa Etária"] = pd.cut(df["Idade"], bins=[0, 18, 30, 50, 100],
                                          labels=["Jovem", "Adulto", "Meia-Idade", "Idoso"])
print(df)
```

```python
# Remover duplicatas
df["Faixa Etária"] = pd.cut(
    df["Idade"],
    bins=[0, 18, 30, 50, 100],
    labels=["Jovem", "Adulto", "Meia-Idade", "Idoso"]
)

print(df)
```

```python
# Montar categoria de salário
df["Faixa Salarial"] = df ["Salário"].map(lambda x: "Baixo"
                                          if x < 1000
                                          else ("Médio" if x < 3500
                                                else "Alto"))

print(df)
```

```python
df.to_excel("dados_tratados.xlsx", index=False)
```
