```python
import pandas as pd

dados = {
    'Nome': ["Jean", "Lúzia", "Sonia", "Tais Helena", "Lucas"],
    'Idade': [25, 45, 67, 15, 18],
    'Naturalidade': ["SP", "AM", "RJ", "MG", "PR",]
}

df = pd.DataFrame(dados)
print(df)
```

```python
# VERIFICAR COLUNA ESPECIFICA
print(df["Nome"])
```

```python
# PUXAR VALOR COM CONDICIONAL
print(df["Idade"]>30)
```

```python
print(df[df["Idade"] > 30])
```

```python
# PUXAR OS REGISTROS DA SONIA
# 1. Filtro df["Nome"]=="Alice"
print(df[
    df["Nome"] == "Alice"
])
```

```python
# df.T

print(df.T)
```

```python
# GERA ESTATISTICAS DESCRITIVAS

df.describe()
```

```python
# Apresenta as primeiras linhas
df.head(2)
```

```python
df["Idade2"] = df["Idade"].apply(lambda x : x + 2)
display(df)
```

```python
display(df["Naturalidade"].value_counts())
```

```python
# ORDENAR DADOS
# ascending = TRUE CRESCENTE
# ascending = False DECRESCENTE

display(df.sort_values("Naturalidade", ascending=False))
```

```python
media_nat = df.groupby("Naturalidade")["Idade"].mean()
display(media_nat)
```
